#### 특성(Feature) 공학과 규제(Regularization : 정형화, 일반화, 규제)


```python
import pandas as pd
```

```python
df = pd.read_csv("../Data/fish.csv")
df.head()
```

```python
fish_full = df.to_numpy()
```

```python
df = pd.read_csv("../Data/fish_weight.csv")
df.head()
```

```python
fish_weight = df.to_numpy()
fish_weight[:5]
```

```python
# Train과 Test
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = \
   train_test_split(
      fish_full,
      fish_weight,
      random_state=42

   )
```

#### sklearn으로 다중회귀 분석하기

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# 다항 특성 만들기

poly = PolynomialFeatures(include_bias=False)
poly.fit([[2,3]])
poly.transform([[2,3]])
```

```python

poly = PolynomialFeatures(include_bias=False) # degree = 2
poly.fit([[2,3]])
poly.transform([[2,3]])
```

```python
poly.get_feature_names_out()
```

----
#### 다향 특성 만들기

```python
train_input.shape
```

```python
# train_input을 다항 특성 만들기
poly = PolynomialFeatures(include_bias=False)
train_poly = poly.fit_transform(train_input)
```

```python
train_input.shape
```

```python
# 컬럼 정보 확인
poly.get_feature_names_out()
```

```python
train_poly[:5]
```

```python
# test_input도 다항특성 만들기
test_poly = poly.transform(test_input)
test_poly.shape
```

#### 다중 회귀 모델 훈련하기

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(train_input, train_target)
print("Train :", lr.score(train_input, train_target))
print("Test :", lr.score(test_input, test_target))
```

```python
# 위의 것보다 결과값이 더 좋아짐

lr = LinearRegression()
lr.fit(train_poly, train_target)
print("Train :", lr.score(train_poly, train_target))
print("Test :", lr.score(test_poly, test_target))
```

----
#### 더 많은 특성 만들기

```python
poly = PolynomialFeatures(include_bias=False, degree=5)
poly.fit(train_input)
train_poly = poly.transform(train_input)
test_poly = poly.transform(test_input)
```

```python
poly.get_feature_names_out()
```

```python
lr.fit(train_poly,train_target)
print("Train : ", lr.score(train_poly, train_target))
print("Test : ", lr.score(test_poly, test_target))
```

```python
train_poly.shape
```

#### 규제
모델이 과적합되게 학습하지 않고 일반성을 가질 수 있도록 규제(정규화)하는 것

```python
# 규제 전에 표준화
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_poly)

train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly)
```

#### Ridge(L2규제)
: 가중치의 제곱으로 규제하는 방법이다

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge()
ridge.fit(train_scaled,train_target )
print("Train : ", ridge.score(train_scaled,train_target))
print("test : ", ridge.score(test_scaled,test_target))
```
