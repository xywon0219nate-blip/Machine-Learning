```python
# Train과 Test로 분리
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target =\
train_test_split(
   iris_dataset['data'],
   iris_dataset['target'],
   random_state=42,
   stratify=iris_dataset['target']
)
```

```python
print(train_input.shape)
print(train_target.shape)
print(test_input.shape)
print(test_target.shape)
```

#### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(train_input, train_target)
```

```python
print(knn.score(train_input, train_target))
print(knn.score(test_input, test_target))
```

```python
import numpy as np
```

```python
for i in range(1,100+1, 2):
   knn = KNeighborsClassifier(n_neighbors=i)
   knn.fit(train_input,train_target)
   pred = knn.predict(test_input)
   result = np.mean(pred == test_target)
   print(f"{i} : {result}")

# 3,5,7,9,11의 결과가 동일하기에 컴퓨터의 입장에선 3을 사용(속도)
```

#### 함수 만들기

```python
def kNNF(i):
   knn = KNeighborsClassifier(n_neighbors=i)
   knn.fit(train_input,train_target)
   pred = knn.predict(test_input)
   result = np.mean(pred == test_target)
   print(f"{i} : {result}")
```

```python
for i in range (1,10+1, 2):
   kNNF(i)
```
