#### scikit learn의 Machine Learning 종류

```python
import pandas as pd
```

```python
csv = pd.read_csv("../Data/iris.csv")
csv.head()
```

```python
# Feature와 Target
iris_data = csv[['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']]
iris_label = csv['Name']
```

```python
# Train과 Test
from sklearn.model_selection import train_test_split

train_data, test_data, train_label, test_label = train_test_split(
   iris_data,
   iris_label,
   random_state=42,
   stratify=iris_label
)
```

----
# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier()
knn_clf.fit(train_data,train_label)
print("knn:", knn_clf.score(test_data,test_label))
```

----
# SVM

```python
from sklearn.svm import SVC
svm_clf = SVC()
svm_clf.fit(train_data,train_label)
print("svm:", svm_clf.score(test_data,test_label))
```

#### Gaussian Process

```python
from sklearn.gaussian_process import GaussianProcessClassifier
gp_clf = GaussianProcessClassifier()
gp_clf.fit(train_data,train_label)
print("GP:", gp_clf.score(test_data,test_label))
```

----
#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier()
dt_clf.fit(train_data, train_label)
print("DT:", dt_clf.score(test_data, test_label))
```

----
#### RandomForesst

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = DecisionTreeClassifier()
rf_clf.fit(train_data,train_label)
print("RandomForest:", rf_clf.score(test_data,test_label))
```

Neural Net

```python
from sklearn.neural_network import MLPClassifier
nn_clf = MLPClassifier(max_iter=1000)
nn_clf.fit(train_data, train_label)
print("NeuralNet:", nn_clf.score(test_data, test_label))
```

----
#### QDA


```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
qda_clf = QuadraticDiscriminantAnalysis()
qda_clf.fit(train_data, train_label)
print("QDA:", qda_clf.score(test_data, test_label))
```

----
#### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb_clf = GaussianNB()
nb_clf.fit(train_data, train_label)
print("NaiveBayes:", nb_clf.score(test_data, test_label))
```
